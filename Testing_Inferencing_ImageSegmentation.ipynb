## Using Single Image

We have trained and evaluated the model. The next step is to see the performance of the model on unknown images.

We are going to test the model on the classes we have trained it on.

```python
import pixellib
from pixellib.instance import custom_segmentation

segment_image = custom_segmentation()
segment_image.inferConfig(num_classes= 2, class_names= ["BG", "Bad Morphological Quality", "Good Morphological Quality"])
segment_image.load_model("/content/drive/Shareddrives/SpermMorphologicalQualityv2/mask_rcnn_models_fin/mask_rcnn_model.056-0.177111.h5")
segment_image.segmentImage("/content/drive/Shareddrives/SpermMorphologicalQualityv2/Inference_test/Normal_Sperm (519).jpg", show_bboxes=True, output_image_name="/content/drive/Shareddrives/SpermMorphologicalQualityv2/Inference_result/bad_out_f1.jpg")
```

```python
import pixellib
from pixellib.instance import custom_segmentation

segment_image = custom_segmentation()
segment_image.inferConfig(num_classes= 2, class_names= ["BG", "Bad Morphological Quality", "Good Morphological Quality"])
segment_image.load_model("/content/drive/Shareddrives/SpermMorphologicalQualityv2/mask_rcnn_models_fin/mask_rcnn_model.056-0.177111.h5")
segment_image.segmentImage("/content/drive/Shareddrives/SpermMorphologicalQualityv2/Inference_test/Normal_Sperm (1).bmp", show_bboxes=True, output_image_name="/content/drive/Shareddrives/SpermMorphologicalQualityv2/Inference_result/good_out_testbmp.jpg")
```

## Video segmentation with a custom model

```python
import pixellib
from pixellib.instance import custom_segmentation

test_video = custom_segmentation()
test_video.inferConfig(num_classes=  2, class_names=["BG", "Bad Morphological Quality", "Good Morphological Quality"])
test_video.load_model("/content/drive/Shareddrives/SpermMorphologicalQualityv2/mask_rcnn_models_fin/mask_rcnn_model.056-0.177111.h5")
test_video.process_video("/content/drive/Shareddrives/SpermMorphologicalQualityv2/Inference_test/Bad7.mp4", show_bboxes = True,  output_video_name="/content/drive/Shareddrives/SpermMorphologicalQualityv2/Inference_result/video_out_b7.mp4", frames_per_second=60)
```

## Live Camera segmentation with a custom model

NOTE: This can only be run using your local computer, not thru Google Colab

```python
import pixellib
from pixellib.instance import custom_segmentation
import cv2

capture = cv2.VideoCapture(0)

segment_camera = custom_segmentation()
segment_camera.inferConfig(num_classes=2, class_names=["BG", "Bad Morphological Quality", "Good Morphological Quality"])
segment_camera.load_model("/content/drive/Shareddrives/SpermMorphologicalQualityv2/mask_rcnn_models_fin/mask_rcnn_model.056-0.177111.h5")
segment_camera.process_camera(capture, frames_per_second= 10, output_video_name="output_video.mp4", show_frames= True,
frame_name= "frame")
```
